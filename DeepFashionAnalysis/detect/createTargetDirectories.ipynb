# Create All directories for file storage

Parse the annotation file and create all directories if not already existing

```python
import os
```

```python
reducedPath = '../fdata_reduce/'
```

```python
annoFile = '../data/Anno/list_bbox.txt'
with open(annoFile,'r') as bbox_file:
    fileList = [line.rstrip('\n').split()[0] for line in bbox_file][2:]
    dirSet = { line.split('/')[1] for line in fileList }
```

```python
if not os.path.exists(reducedPath): os.mkdir(reducedPath)
for d in dirSet:
    dirA = os.path.join(reducedPath,d.split('_')[-1])
    dirB = os.path.join(dirA,d)
    if not os.path.exists(dirA): os.mkdir(dirA)
    if not os.path.exists(dirB): os.mkdir(dirB)
    
```
