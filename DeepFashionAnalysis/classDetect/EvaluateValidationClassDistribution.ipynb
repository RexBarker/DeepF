# Evaluate class distribution, validation set
- summary of the classes as used for the image validation

```python
import os
from glob import glob
import matplotlib.pyplot as plt
```

```python
valDir = '../fashion_data_class/val/'
generalDir = '../fashion_data/'
valFiles = glob(valDir + '**/*.jpg',recursive=True)
```

```python
cl_names = [l for l in os.listdir(valDir) if not l.startswith('.') ]
cl_count = {}
for imgset in ('test','train','val'):
    cl_count[imgset] = { cl:len(glob(generalDir + imgset + '/'+ cl + '/**/*.jpg',recursive=True)) 
                        for cl in cl_names}
```

```python
cl_count
```

```python
imgsetcolors = {'train':'orange', 'test':'cyan', 'val':'violet'}
for imgset,cl_c in cl_count.items():
    cl_sort= sorted(cl_c.items(),key=lambda kv: kv[1])
    names = [ k for k,v in cl_sort]
    vals = [v for k,v in cl_sort]

    fig = plt.figure(figsize=(8,6))
    plt.bar(range(len(vals)),vals,color=imgsetcolors[imgset])
    for x in range(len(vals)): plt.text(x=x,y=vals[x]+0.1*max(vals),s=str(vals[x]),rotation='vertical',fontsize=12)
    plt.xticks(range(len(vals)),names,rotation='vertical',fontsize=14)
    plt.yticks(fontsize=14)
    plt.grid(color='grey',axis='y',linestyle='--', linewidth=0.5)
    plt.title(f"Class distribution 'Upper'\n'{imgset}' Set",fontsize=20,fontweight='bold')

    plt.show()
```
